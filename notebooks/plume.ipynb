```python
%matplotlib inline
from IPython.display import Image
import numpy as np
from fatiando.gravmag.blocky import Gravity, sow
from fatiando.gravmag import tesseroid
from fatiando import gridder, utils
from fatiando.mesher import Tesseroid, TesseroidMesh, vremove
from fatiando.vis import mpl, myv
imgdir = 'fig'
```

```python
props = {'density':-50}
model = [Tesseroid(-156.5, -154.5, 19, 21, -100000, -300000, props),
         Tesseroid(-156, -155, 19.5, 20.5, -300000, -1000000, props)]
```

```python
def snapshot(scene, fname):    
    scene.scene.camera.position = [-4845883.9070980623, -3297812.4602981922, 3510960.3428494777]
    scene.scene.camera.focal_point = [-5008643.0944703668, -2310303.8069823612, 2046307.549933386]
    scene.scene.camera.view_angle = 30.0
    scene.scene.camera.view_up = [-0.99193521467119017, -0.12390409176946064, 0.026689060268301784]
    scene.scene.camera.clipping_range = [1045425.4948085039, 2748894.2133707912]
    scene.scene.camera.compute_view_plane_normal()
    scene.scene.render()
    myv.savefig('{}/{}.png'.format(imgdir, fname))
```

```python
scene = myv.figure(zdown=False, size=(400, 700))
myv.tesseroids(model, linewidth=1, color=(1, 0, 0))
snapshot(scene, 'plume-model')
myv.show()
```

```python
Image(filename='plume-model.png')
```

```python
bounds = [-158.5, -152.5, 17, 23, 0, -1500000]
area = [-170, -140, 5, 35]
shape = (50, 50)
lon, lat, height = gridder.regular(area, shape, z=250000)
gz = utils.contaminate(tesseroid.gz(lon, lat, height, model), 0.1, seed=0)
gzz = utils.contaminate(tesseroid.gzz(lon, lat, height, model), 0.001, seed=0)
```

```python
bm = mpl.basemap(area, 'merc', resolution='h')
for f, units, fname in zip([gz, gzz], ['mGal', r'$E\"otv\"os$'], ['plume-gz.png', 'plume-gzz.png']):
    mpl.figure()
    mpl.contourf(lon, lat, f, shape, 50, basemap=bm)
    mpl.colorbar().set_label(units)
    mpl.draw_geolines(area, 10, 10, basemap=bm)
    bm.drawcoastlines()
    mpl.savefig('{}/{}'.format(imgdir, fname), dpi=200)
```

```python
def plot(solver, data, fname):
    predicted = solver.predicted()
    if len(np.shape(predicted)) == 1:
        predicted = [predicted]
    residuals = solver.residuals()
    if len(np.shape(residuals)) == 1:
        residuals = [residuals]
    for i, d, p, r in zip(range(len(data)), data, predicted, residuals):
        mpl.figure()
        bm.drawcoastlines()
        bm.drawmapboundary()
        mpl.draw_geolines(area, 10, 10, basemap=bm)
        levels = mpl.contour(lon, lat, d, shape, 10, basemap=bm, color='b', label='observed')
        mpl.contour(lon, lat, p, shape, levels, basemap=bm, color='r', label='predicted')
        mpl.legend(fancybox=True, shadow=True)
        mpl.savefig('{}/plume-fit-{}-{}.png'.format(imgdir, fname, i), dpi=200)
        mpl.figure()
        mpl.hist(r, bins=20)
        mpl.title("Residuals")
        mpl.savefig('{}/plume-residuals-{}-{}.png'.format(imgdir, fname, i), dpi=200)
    mesh.addprop('density', solver.estimate_)
    scene = myv.figure(zdown=False, size=(400, 700))
    myv.tesseroids(model, linewidth=1, color=(1, 0, 0), opacity=0.4)
    myv.tesseroids(seeds, linewidth=1)
    myv.tesseroids(vremove(0, 'density', mesh), linewidth=1)
    snapshot(scene, 'plume-estimate-' + fname)
    myv.show()
```

```python
mesh = TesseroidMesh(bounds, (30, 30, 30))
mesh.dims
```

```python
seeds = sow([[-155.5, 20, -100000 + 1, props]], mesh)

scene = myv.figure(zdown=False, size=(400, 700))
myv.tesseroids(model, linewidth=1, color=(1, 0, 0), opacity=0.4)
myv.tesseroids(seeds, linewidth=1)
snapshot(scene, 'plume-seeds')
myv.show()
```

```python
Image(filename='{}/plume-seeds.png'.format(imgdir))
```

```python
solver = Gravity(lon, lat, height, gzz, mesh, field='gzz')
```

```python
solver.config('planting', seeds=seeds, compactness=0.0005, threshold=0.0001).fit()
```

```python
plot(solver, [gzz], 'single')
```

```python
Image(filename='{}/plume-estimate-single.png'.format(imgdir))
```

```python
joint = 5*solver + Gravity(lon, lat, height, gz, mesh, field='gz')
```

```python
joint.config('planting', seeds=seeds, compactness=0.015, threshold=0.0001).fit()
```

```python
plot(joint, [gzz, gz], 'joint')
```
