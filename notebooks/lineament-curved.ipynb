```python
%matplotlib inline
import numpy as np
from fatiando import gridder, utils
from fatiando.gravmag import tesseroid
from fatiando.gravmag.blocky import sow, Gravity
from fatiando.mesher import Tesseroid, TesseroidMesh, vremove
from fatiando.vis import mpl, myv
```

# Make a model

```python
props = {'density':300}
model = ([Tesseroid(6, 6.5, 18, 25, -2000, -10000, props)]
         + [Tesseroid(6 - i*0.25 - 0.25, 6 - i*0.25 + 0.25, 18 - (i+1)*0.5, 18 - i*0.5, -2000, -10000, props)
            for i in range(6)]
         + [Tesseroid(4, 4.5, 16, 21, -2000, -10000, props)]
         + [Tesseroid(4 - i*0.25 - 0.25, 4 - i*0.25 + 0.25, 16 - (i+1)*0.5, 16 - i*0.5, -2000, -10000, props)
            for i in range(4)]
         )
```

```python
bounds = [0, 10, 10, 30, 0, -20000]
```

```python
myv.figure(zdown=False)
myv.tesseroids(model, edges=False, color=(1, 0, 0), scale=(1, 1, 5))
myv.show()
```

```python
shape = (50, 50)
area = [0, 10, 10, 30]
lon, lat, height = gridder.regular(area, shape, z=20000)
gz = utils.contaminate(tesseroid.gz(lon, lat, height, model), 0.01, percent=True, seed=0)
gzz = utils.contaminate(tesseroid.gzz(lon, lat, height, model), 0.01, percent=True, seed=0)
height2 = height + 100000
gzz2 = utils.contaminate(tesseroid.gzz(lon, lat, height2, model), 0.01, percent=True, seed=0)
height3 = height + 250000
gzz3 = utils.contaminate(tesseroid.gzz(lon, lat, height3, model), 0.01, percent=True, seed=0)
```

```python
bm = mpl.basemap(area, 'merc')
for f in [gz, gzz, gzz2, gzz3]:
    mpl.figure()
    mpl.contourf(lon, lat, f, shape, 50, basemap=bm)
    mpl.colorbar()
    mpl.draw_geolines(area, 5, 10, basemap=bm)
```

```python
def plot(solver, data):
    predicted = solver.predicted()
    if len(np.shape(predicted)) == 1:
        predicted = [predicted]
    residuals = solver.residuals()
    if len(np.shape(residuals)) == 1:
        residuals = [residuals]
    for d, p, r in zip(data, predicted, residuals):
        mpl.figure(figsize=(12, 6))
        mpl.subplot(1, 3, 1)
        mpl.title("Observed")
        l = mpl.contourf(lon, lat, d, shape, 30, basemap=bm)
        mpl.colorbar()
        mpl.draw_geolines(area, 5, 10, basemap=bm)
        mpl.subplot(1, 3, 2)
        mpl.title("Predicted")
        mpl.contourf(lon, lat, p, shape, l, basemap=bm)
        mpl.colorbar()
        mpl.draw_geolines(area, 5, 10, basemap=bm)
        mpl.subplot(1, 3, 3)
        mpl.title('Residuals')
        mpl.hist(r, bins=20)
    mesh.addprop('density', solver.estimate_)
    myv.figure()
    scale = (1, 1, 5)
    myv.tesseroids(model, opacity=0.4, edges=False, scale=scale)
    myv.tesseroids(vremove(0, 'density', mesh), color=(1, 0, 0), scale=scale, linewidth=1)
    myv.tesseroids(seeds, 'density', scale=scale)
    myv.show()
```

```python
mesh = TesseroidMesh(bounds, (10, 40, 100))
print mesh.dims
```

```python
seeds = sow([[6.25, 20, -2000 + 1, props],
             [4.25, 18, -2000 + 1, props]], mesh)

scale = (1, 1, 5)
myv.figure(zdown=False)
myv.tesseroids(model, edges=False, opacity=0.2, scale=scale)
myv.tesseroids(seeds, linewidth=1, color=(1, 0, 0), scale=scale)
myv.show()
```

```python
solver_gz = Gravity(lon, lat, height, gz, mesh)
```

```python
solver_gz.config('planting', seeds=seeds, compactness=10, threshold=0.0001).fit()
```

```python
plot(solver_gz, [gz])
```

```python
solver_gzz = Gravity(lon, lat, height, gzz, mesh, field='gzz')
```

```python
solver_gzz.config('planting', seeds=seeds, compactness=1, threshold=0.0001).fit()
```

```python
plot(solver_gzz, [gzz])
```
