```python
%matplotlib inline
from IPython.display import Image
import numpy as np
import cPickle as pickle
from fatiando.gravmag import tesseroid
from fatiando.gravmag.blocky import Gravity, sow
from fatiando.mesher import TesseroidMesh, Tesseroid, vremove
from fatiando import gridder
from fatiando.utils import contaminate
from fatiando.vis import mpl, myv
```

# Make a tesseroid model of a subducting plate

```python
w, e, s, n, top, bottom = -75, -60, -50, -20, -100000, -2000000
wests = np.linspace(w, e, 6)
souths = np.linspace(s, n, 20)
dlon = wests[1] - wests[0]
print dlon
bottoms = np.linspace(bottom, top, len(wests))[::-1]
model = [Tesseroid(wests[i] - 0.5*(1 + 0.2*i)*dlon, wests[i+1], souths[j], souths[j+1], bottoms[i], bottoms[i+1], {'density':-80})
         for i in range(len(wests)-1) for j in range(len(souths)-1)]
print max(t.top for t in model)
```

```python
scene = myv.figure(zdown=False, size=(800, 600))
myv.tesseroids(model, 'density', linewidth=3)
myv.earth(opacity=0.4)
myv.core(opacity=1)
myv.continents(linewidth=1.5)
myv.show()
```

# Simulate satellite data

```python
bounds = [-90, -50, -60, 0, 0, -3000000]
shape = (50, 50)
area = [-110, -30, -70, 10]
lons, lats, heights = gridder.regular(area, shape, z=250000)
potential = contaminate(tesseroid.potential(lons, lats, heights, model), 0.01, percent=True, seed=0)
gz = contaminate(tesseroid.gz(lons, lats, heights, model), 0.01, percent=True, seed=0)
gxx = contaminate(tesseroid.gxx(lons, lats, heights, model), 0.01, percent=True, seed=0)
gxy = contaminate(tesseroid.gxy(lons, lats, heights, model), 0.01, percent=True, seed=0)
gxz = contaminate(tesseroid.gxz(lons, lats, heights, model), 0.01, percent=True, seed=0)
gyy = contaminate(tesseroid.gyy(lons, lats, heights, model), 0.01, percent=True, seed=0)
gyz = contaminate(tesseroid.gyz(lons, lats, heights, model), 0.01, percent=True, seed=0)
gzz = contaminate(tesseroid.gzz(lons, lats, heights, model), 0.01, percent=True, seed=0)
```

```python
#bm = mpl.basemap([-140, -30, -70, 20], 'cass')
bm = mpl.basemap(area, 'ortho')
for i, f in enumerate([potential, gz, gzz]):
    mpl.figure()
    bm.drawcoastlines()
    bm.drawmapboundary()
    mpl.contourf(lons, lats, f, shape, 50, basemap=bm)
    mpl.colorbar()
```

# Make some utility functions

```python
def plots(solver, data):
    predicted = solver.predicted()
    if len(np.shape(predicted)) == 1:
        predicted = [predicted]
    residuals = solver.residuals()
    if len(np.shape(residuals)) == 1:
        residuals = [residuals]
    for d, p, r in zip(data, predicted, residuals):
        mpl.figure(figsize=(12, 5))
        mpl.subplot(1, 2, 1)
        bm.drawcoastlines()
        bm.drawmapboundary()
        levels = mpl.contour(lons, lats, d, shape, 10, basemap=bm, color='b')
        mpl.contour(lons, lats, p, shape, levels, basemap=bm, color='r')
        mpl.title("Red = predicted, blue = observed")
        mpl.subplot(1, 2, 2)
        mpl.hist(r, bins=20)
        mpl.title("Residuals")
    scene = myv.figure(zdown=False, size=(800, 600))
    myv.tesseroids(model, edges=False, color=(0,0,1), opacity=0.2)
    myv.tesseroids(seeds, prop='density', edges=True, color=(0, 0, 0), linewidth=1)
    myv.tesseroids(vremove(0, 'density', mesh), prop='density', edges=True, color=(1, 0, 0), linewidth=1)
    #myv.earth(opacity=0.4)
    #myv.core(opacity=1)
    #myv.continents(linewidth=1.5)
    myv.show()
```

```python
def spine(mesh):
    nz, ny, nx = mesh.shape
    locations = []
    for i, layer in enumerate(mesh.layers()):
        if i == nz - 1:
            break
        nonzero = vremove(0, 'density', layer)
        if len(nonzero) > 10:
            centerofmass = sum(p.center() for p in nonzero)/len(nonzero)
            locations.append(centerofmass.tolist() + [props])
    return locations
```

# Run the inversion using gzz

```python
mesh = TesseroidMesh(bounds, (30, 30, 40))
props = {'density':-80}
mesh.dims
```

```python
original = [[-75, -35, -90000, props]]
locations = original
```

```python
seeds = sow(locations, mesh)

myv.figure(zdown=False)
myv.tesseroids(seeds, color=(1, 0, 0))
myv.tesseroids(model, opacity=0.4)
myv.show()
```

```python
solver = (Gravity(lons, lats, heights, gzz, mesh, field='gzz')
          + Gravity(lons, lats, heights, gxx, mesh, field='gxx')
          + Gravity(lons, lats, heights, gyy, mesh, field='gyy')
          )
```

```python
solver.config('planting', seeds=seeds, compactness=0.01, threshold=0.001).fit()
mesh.addprop('density', solver.estimate_)
```

```python
plots(solver, [gzz, gxx, gyy])
```

```python
locations = spine(mesh)
```

```python
seeds = sow(locations, mesh)

myv.figure(zdown=False)
myv.tesseroids(seeds, color=(1, 0, 0))
myv.tesseroids(model, opacity=0.4)
myv.show()
```

```python
solver.config('planting', seeds=seeds, compactness=0.02, threshold=0.001).fit()
mesh.addprop('density', solver.estimate_)
```
