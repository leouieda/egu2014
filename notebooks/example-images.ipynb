```python
%matplotlib inline
from IPython.display import Image
from fatiando.mesher import Tesseroid, Prism, PrismMesh
from fatiando.vis import myv
imgdir = 'fig'
```

```python
prism = PrismMesh((0, 3, 50, 80, 0, 3), (1, 10, 1))
```

```python
scene = myv.figure(size=(400, 700))
myv.prisms(prism, linewidth=1, color=(1, 0, 0))
scene.scene.camera.position = [-14.606171036362635, 27.648658764332662, -32.050315806590909]
scene.scene.camera.focal_point = [2.1213482538808535, 63.839227123110042, 2.4408353381475925]
scene.scene.camera.view_angle = 30.0
scene.scene.camera.view_up = [0.018759894580073794, 0.68522743035167366, -0.72808751881143186]
scene.scene.camera.clipping_range = [28.783459833581823, 82.93840547060762]
scene.scene.camera.compute_view_plane_normal()
scene.scene.render()
myv.savefig('{}/example-prism.png'.format(imgdir))
myv.show()
```

```python
Image(filename='{}/example-prism.png'.format(imgdir))
```

```python
tess = Tesseroid(0, 5, 50, 70, 200e3, 0).split(1, 10, 1)
```

```python
scene = myv.figure(zdown=False, size=(400, 700))
myv.tesseroids(tess, linewidth=1, color=(1, 0, 0))
myv.earth(opacity=1)
myv.continents(resolution=1, linewidth=2)
scene.scene.camera.position = [7029483.6544555426, 1573216.4191872336, 6102119.5609793188]
scene.scene.camera.focal_point = [-1424632.9000719553, -1822044.5943514351, 4336966.9442494074]
scene.scene.camera.view_angle = 30.0
scene.scene.camera.view_up = [-0.18130384179602471, -0.058910866847267859, 0.98166105490505506]
scene.scene.camera.clipping_range = [30780.487858078024, 30780487.858078025]
scene.scene.camera.compute_view_plane_normal()
scene.scene.render()
myv.savefig('{}/example-tesseroid.png'.format(imgdir), magnification=1.5)
myv.show()
```
