```python
%matplotlib inline
from IPython.display import Image
import numpy as np
from fatiando import gridder, utils
from fatiando.gravmag import tesseroid
from fatiando.gravmag.blocky import sow, Gravity
from fatiando.mesher import Tesseroid, TesseroidMesh, vremove
from fatiando.vis import mpl, myv
imgdir = 'fig'
```

# Make a model

```python
props = {'density':300}
model = (Tesseroid(6, 6.5, 18, 25, -2000, -10000, props).split(1, 5, 1)
         + [Tesseroid(6 - i*0.25 - 0.25, 6 - i*0.25 + 0.25, 18 - (i+1)*0.5, 18 - i*0.5, -2000, -10000, props)
            for i in range(6)]
         + Tesseroid(4, 4.5, 16, 21, -2000, -10000, props).split(1, 5, 1)
         + [Tesseroid(4 - i*0.25 - 0.25, 4 - i*0.25 + 0.25, 16 - (i+1)*0.5, 16 - i*0.5, -2000, -10000, props)
            for i in range(4)])
```

```python
bounds = [0, 10, 10, 30, 0, -20000]
```

```python
def snapshot(scene, fname):
    scene.scene.camera.position = [6666184.0617848355, 179399.26276113046, 1233440.0080897624]
    scene.scene.camera.focal_point = [5604900.5651898468, 648133.15418051148, 2178202.0930754482]
    scene.scene.camera.view_angle = 30.0
    scene.scene.camera.view_up = [0.70487758681038615, 0.31072965235262112, 0.63764776386433442]
    scene.scene.camera.clipping_range = [2672.6558134350335, 2672655.8134350334]
    scene.scene.camera.compute_view_plane_normal()
    scene.scene.render()
    myv.savefig('{}/{}.png'.format(imgdir, fname))
    
scale = (1, 1, 5)
size = (600, 700)
```

```python
scene = myv.figure(zdown=False, size=size)
myv.tesseroids(model, edges=True, color=(1, 0, 0), linewidth=1, scale=scale)
snapshot(scene, 'curved-model')
myv.show()
```

```python
Image(filename='{}/curved-model.png'.format(imgdir))
```

```python
shape = (50, 50)
area = [0, 10, 10, 30]
lon, lat, height = gridder.regular(area, shape, z=20000)
gzz = utils.contaminate(tesseroid.gzz(lon, lat, height, model), 0.01, percent=True, seed=0)
height2 = height + 100000
gzz2 = utils.contaminate(tesseroid.gzz(lon, lat, height2, model), 0.01, percent=True, seed=0)
height3 = height + 250000
gzz3 = utils.contaminate(tesseroid.gzz(lon, lat, height3, model), 0.01, percent=True, seed=0)
```

```python
bm = mpl.basemap(area, 'merc')
```

```python
for i, f in enumerate([gzz, gzz2, gzz3]):
    mpl.figure(figsize=(3, 4))
    mpl.subplots_adjust(left=-0.2)
    mpl.contourf(lon, lat, f, shape, 50, basemap=bm)
    mpl.colorbar(pad=0.01).set_label(r'$E\"otv\"os$')
    mpl.draw_geolines(area, 5, 5, basemap=bm)
    mpl.savefig('{}/curved-height-{}.png'.format(imgdir, i), dpi=200)
```

```python
def plot(solver, data, fname):
    predicted = solver.predicted()
    residuals = solver.residuals()
    
    mpl.figure()
    mpl.subplots_adjust(left=0.2, wspace=0, right=0.87)
    mpl.subplot(1, 2, 1)
    levels = mpl.contourf(lon, lat, data, shape, 30, basemap=bm)
    mpl.draw_geolines(area, 5, 5, basemap=bm)
    mpl.subplot(1, 2, 2)
    mpl.contourf(lon, lat, predicted, shape, levels, basemap=bm)
    mpl.colorbar(pad=0).set_label(r'$E\"otv\"os$')
    mpl.draw_geolines(area, 5, 5, basemap=bm)
    mpl.savefig('{}/curved-fit-{}.png'.format(imgdir, fname), dpi=200)
    
    mpl.figure(figsize=(3, 2))
    mpl.subplots_adjust(left=0.15)
    mpl.title('Residuals')
    mpl.hist(residuals, bins=20)
    ax = mpl.gca()
    ax.set_xticks(ax.get_xticks()[::2])
    mpl.savefig('{}/curved-residuals-{}.png'.format(imgdir, fname), dpi=200)
    
    mesh.addprop('density', solver.estimate_)    
    scene = myv.figure(zdown=False, size=size)
    myv.tesseroids(model, style='wireframe', color=(1, 0, 0), linewidth=5, scale=scale)
    myv.tesseroids(vremove(0, 'density', mesh), scale=scale, linewidth=1)
    snapshot(scene, 'curved-estimate-' + fname)
    myv.show()
```

```python
mesh = TesseroidMesh(bounds, (10, 40, 100))
print mesh.dims
```

```python
seeds = sow([[6.25, 20, -2000 + 1, props],
             [4.25, 18, -2000 + 1, props]], mesh)
```

```python
scene = myv.figure(zdown=False, size=size)
myv.tesseroids(model, style='wireframe', color=(0, 0, 0), linewidth=1, scale=scale)
myv.tesseroids(model, edges=False, color=(1, 0, 0), opacity=0.6, scale=scale)
myv.tesseroids(seeds, linewidth=1, scale=scale)
snapshot(scene, 'curved-seeds')
myv.show()
```

```python
Image(filename='{}/curved-seeds.png'.format(imgdir))
```

```python
solver_gzz = Gravity(lon, lat, height, gzz, mesh, field='gzz')
```

```python
solver_gzz.config('planting', seeds=seeds, compactness=1, threshold=0.0001).fit()
```

```python
plot(solver_gzz, gzz, 'height-0')
```
