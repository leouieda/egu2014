```python
%matplotlib inline
import numpy as np
from fatiando.gravmag import harvester, tesseroid
from fatiando.mesher import Tesseroid, TesseroidMesh, vremove
from fatiando import gridder, utils
from fatiando.vis import mpl, myv
```

```python
w, e, s, n, top, bottom = -70, -60, -50, -20, -10000, -300000
wests = np.linspace(w, e, 20)
souths = np.linspace(s, n, 20)
dlon = wests[1] - wests[0]
bottoms = np.linspace(bottom, top, len(wests))[::-1]
density = 2900 - 3400
subduction_model = [Tesseroid(wests[i] - 0.7*dlon, wests[i+1], souths[j], souths[j+1], bottoms[i], bottoms[i+1], {'density':-200})
                    for i in range(len(wests)-1) for j in range(len(souths)-1)]
print abs(bottoms[0])
```

```python
scene = myv.figure(zdown=False, size=(800,600))
myv.tesseroids(subduction_model, 'density')
myv.earth(opacity=0.4)
myv.core(opacity=1)
myv.continents(linewidth=1.5)
#subduction_snapshots(scene, 'subduction-model')
myv.show()
```

```python
bounds = [-90, -40, -60, -10, 0, -800000]
shape = (50, 50)
lons, lats, heights = gridder.regular(bounds[:4], shape, z=4000)
gz = utils.contaminate(tesseroid.gz(lons, lats, heights, subduction_model), 0.1)

mpl.figure(figsize=(12, 10))
bm = mpl.basemap(bounds[:4], 'ortho')
bm.drawcoastlines()
bm.drawmapboundary()
mpl.contourf(lons, lats, gz, shape, 50, basemap=bm)
cb = mpl.colorbar()
cb.set_label('mGal')
#mpl.savefig('images/subduction-gravity.png', dpi=500)
```

```python
mesh = TesseroidMesh(bounds, (20, 20, 80))
props = subduction_model[0].props

seeds = harvester.sow([[-69, -45, -5000, props],
                       [-69, -40, -5000, props],
                       [-69, -35, -5000, props],
                       [-69, -30, -5000, props],
                       [-69, -25, -5000, props]], mesh)
```

```python
mesh = TesseroidMesh(bounds, (40, 20, 100))
props = subduction_model[0].props
seeds = harvester.sow([[-70, -35, -10000, props],
                       [-68, -35, -40000, props],
                       [-66, -35, -100000, props],
                       [-64, -35, -160000, props],
                       [-63, -35, -180000, props],
                       [-61, -35, -240000, props],], mesh)
```

```python
scene = myv.figure(zdown=False, size=(800, 600))
scale = (1, 1, 1)
myv.tesseroids(subduction_model, edges=False, opacity=0.4, color=(0.8,0.8,0.8), scale=scale)
myv.tesseroids(seeds, scale=scale)
#myv.earth(opacity=0.4)
#myv.core(opacity=1)
#myv.continents(linewidth=1.5)
#subduction_snapshots(scene, 'subduction-seed')
myv.show()
```

```python
data = [harvester.Gz(lons, lats, heights, gz, meshtype='tesseroid')]
estimate, predicted = harvester.harvest(data, seeds, mesh, compactness=0.01, threshold=0.000001)
mesh.addprop('density', estimate['density'])
subduction_estimate = vremove(0, 'density', mesh)
print len(subduction_estimate)
```

```python
mpl.figure(figsize=(12,10))
bm = mpl.basemap(bounds[:4], 'ortho')
bm.drawcoastlines()
bm.drawmapboundary()
levels = mpl.contour(lons, lats, gz, shape, 10, basemap=bm)
mpl.contour(lons, lats, predicted[0], shape, levels, basemap=bm, color='r')
#mpl.savefig('images/subdution-predicted.png', dpi=500)
```
