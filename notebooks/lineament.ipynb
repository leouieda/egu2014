```python
%matplotlib inline
import numpy as np
from fatiando import gridder, utils
from fatiando.gravmag import tesseroid
from fatiando.gravmag.blocky import sow, Gravity
from fatiando.mesher import Tesseroid, TesseroidMesh, vremove
from fatiando.vis import mpl, myv
```

# Make a model

```python
model = Tesseroid(0, 0.5, 15, 25, -2000, -10000, {'density':300}).split(1, 20, 1)
```

```python
bounds = [-5, 5, 10, 30, 0, -20000]
```

```python
myv.figure(zdown=False)
myv.tesseroids(model, edges=False, color=(1, 0, 0), scale=(1, 1, 5))
myv.show()
```

```python
shape = (100, 100)
area = [-5, 5, 10, 30]
lon, lat, height = gridder.regular(area, shape, z=20000)
gz = utils.contaminate(tesseroid.gz(lon, lat, height, model), 0.01, percent=True, seed=0)
gzz = utils.contaminate(tesseroid.gzz(lon, lat, height, model), 0.01, percent=True, seed=0)
```

```python
bm = mpl.basemap(area, 'merc')
for f in [gz, gzz]:
    mpl.figure()
    mpl.contourf(lon, lat, f, shape, 50, basemap=bm)
    mpl.colorbar()
    mpl.draw_geolines(area, 5, 10, basemap=bm)
```

```python
def plot(solver, data):
    predicted = solver.predicted()
    if len(np.shape(predicted)) == 1:
        predicted = [predicted]
    residuals = solver.residuals()
    if len(np.shape(residuals)) == 1:
        residuals = [residuals]
    for d, p, r in zip(data, predicted, residuals):
        mpl.figure(figsize=(12, 6))
        mpl.subplot(1, 3, 1)
        mpl.title("Observed")
        l = mpl.contourf(lon, lat, d, shape, 30, basemap=bm)
        mpl.colorbar()
        mpl.draw_geolines(area, 5, 10, basemap=bm)
        mpl.subplot(1, 3, 2)
        mpl.title("Predicted")
        mpl.contourf(lon, lat, p, shape, l, basemap=bm)
        mpl.colorbar()
        mpl.draw_geolines(area, 5, 10, basemap=bm)
        mpl.subplot(1, 3, 3)
        mpl.title('Residuals')
        mpl.hist(r, bins=20)
    myv.figure()
    scale = (1, 1, 5)
    myv.tesseroids(model, opacity=0.4, edges=False, scale=scale)
    myv.tesseroids(vremove(0, 'density', mesh), color=(1, 0, 0), scale=scale, linewidth=1)
    myv.tesseroids(seeds, 'density', scale=scale)
    myv.show()
```

```python
mesh = TesseroidMesh(bounds, (10, 40, 100))
print mesh.dims
```

```python
seeds = sow([[0.25, 20, -2000 + 1, {'density':300}]], mesh)

scale = (1, 1, 5)
myv.figure(zdown=False)
myv.tesseroids(model, edges=False, opacity=0.2, scale=scale)
myv.tesseroids(seeds, linewidth=1, color=(1, 0, 0), scale=scale)
myv.show()
```

```python
solver = Gravity(lon, lat, height, gz, mesh)
```

```python
solver.config('planting', seeds=seeds, compactness=10, threshold=0.0001).fit()
mesh.addprop('density', solver.estimate_)
```

```python
plot(solver, [gz])
```

```python
solver = Gravity(lon, lat, height, gz, mesh) + Gravity(lon, lat, height, gzz, mesh, field='gzz')
```

```python
solver.config('planting', seeds=seeds, compactness=10, threshold=0.0001).fit()
mesh.addprop('density', solver.estimate_)
```

```python
plot(solver, [gz, gzz])
```
