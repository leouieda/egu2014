```python
%matplotlib inline
from IPython.display import Image
import numpy as np
from fatiando import gridder, utils
from fatiando.gravmag import tesseroid
from fatiando.gravmag.blocky import sow, Gravity
from fatiando.mesher import Tesseroid, TesseroidMesh, vremove
from fatiando.vis import mpl, myv
imgdir = 'fig'
```

# Make a model

```python
props = {'density':200}
model = (Tesseroid(-53, -48, -25, -15, -30e3, -35e3, props).split(5, 5, 1)
         + Tesseroid(-52, -49, -24,  -17, -35e3, -40e3, props).split(3, 3, 1)
         + Tesseroid(-52, -49, -23,  -20, -40e3, -45e3, props).split(3, 3, 1)
         )
```

```python
bounds = [-60, -40, -30, -10, -30e3, -70e3]
```

```python
def snapshot(scene, fname):
    scene.scene.camera.position = [4068226.0831730124, -3680321.6509752572, -2305621.7994582746]
    scene.scene.camera.focal_point = [3384183.1384716178, -4016232.5035612206, -1974105.4934253355]
    scene.scene.camera.view_angle = 30.0
    scene.scene.camera.view_up = [0.34056282493391049, -0.91343431293636346, -0.22283338668949151]
    scene.scene.camera.clipping_range = [2344.8378426521012, 2344837.8426521011]
    scene.scene.camera.compute_view_plane_normal()
    scene.scene.render()
    myv.savefig('{}/{}.png'.format(imgdir, fname))

size = (1200, 400)
scale = (1, 1, 20)
```

```python
scene = myv.figure(zdown=False, size=size)
myv.tesseroids(model, edges=True, color=(1, 0, 0), scale=scale, linewidth=1)
snapshot(scene, 'underplate-model')
myv.show()
```

```python
Image(filename='{}/underplate-model.png'.format(imgdir))
```

```python
shape = (50, 50)
area = bounds[:4]
lon, lat, height = gridder.regular(area, shape, z=250e3)
gzz = utils.contaminate(tesseroid.gzz(lon, lat, height, model), 0.001, seed=0)
```

```python
bm = mpl.basemap(area, 'merc')
```

```python
mpl.figure(figsize=(5, 4))
mpl.subplots_adjust(left=-0.2)
mpl.contourf(lon, lat, gzz, shape, 50, basemap=bm)
mpl.colorbar(pad=0.01).set_label(r'$E\"otv\"os$')
mpl.draw_geolines(area, 5, 5, basemap=bm)
mpl.savefig('{}/underplate-gzz.png'.format(imgdir), dpi=200)
```

```python
def plot(solver, data, fname):
    predicted = solver.predicted()
    residuals = solver.residuals()
    
    mpl.figure(figsize=(5, 4))
    mpl.subplots_adjust(left=-0.2)
    levels = mpl.contour(lon, lat, data, shape, 10, color='b', label='observed', basemap=bm)
    mpl.contour(lon, lat, predicted, shape, levels, color='r', label='predicted', basemap=bm)    
    mpl.legend(fancybox=True, shadow=True)
    mpl.draw_geolines(area, 5, 5, basemap=bm)
    mpl.savefig('{}/underplate-fit-{}.png'.format(imgdir, fname), dpi=200)
    
    mpl.figure(figsize=(3, 2))
    mpl.subplots_adjust(left=0.15)
    mpl.title('Residuals')
    mpl.hist(residuals, bins=20)
    ax = mpl.gca()
    ax.set_xticks(ax.get_xticks()[::2])
    mpl.savefig('{}/underplate-residuals-{}.png'.format(imgdir, fname), dpi=200)
    
    mesh.addprop('density', solver.estimate_)    
    scene = myv.figure(zdown=False, size=size)
    myv.tesseroids(model, style='wireframe', color=(1, 0, 0), linewidth=5, scale=scale)
    myv.tesseroids(vremove(0, 'density', mesh), scale=scale, linewidth=1)
    snapshot(scene, 'underplate-estimate-' + fname)
    myv.show()
```

```python
mesh = TesseroidMesh(bounds, (40, 40, 40))
print mesh.dims
```

```python
seeds = sow([[-50, -21, -30e3, props]], mesh)
```

```python
scene = myv.figure(zdown=False, size=size)
myv.tesseroids(model, style='wireframe', color=(0, 0, 0), linewidth=1, scale=scale)
myv.tesseroids(model, edges=False, color=(1, 0, 0), opacity=0.5, scale=scale)
myv.tesseroids(seeds, linewidth=1, scale=scale)
snapshot(scene, 'underplate-seeds')
myv.show()
```

```python
Image(filename='{}/underplate-seeds.png'.format(imgdir))
```

```python
solver = Gravity(lon, lat, height, gzz, mesh, field='gzz')
```

```python
solver.config('planting', seeds=seeds, compactness=0.01, threshold=0.0001).fit()
```

```python
plot(solver, gzz, 'correct')
```

```python
Image(filename='{}/underplate-estimate-correct.png'.format(imgdir))
```
